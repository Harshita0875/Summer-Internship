```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder \
    .appName("SalesDataFrame") \
    .master("local[*]") \
    .getOrCreate()

print("Spark Started")
```

```python
df = spark.read.csv(
    "/workspace/sales.csv",
    header=True,
    inferSchema=True
)

df.show()
```

```python
import os

os.makedirs("/workspace/output", exist_ok=True)

print("Output folder created")
```

```python
sorted_df = df.orderBy(
    col(" sales ").desc()
)

print("===== Products Sorted By Sales =====")

sorted_df.show()
sorted_df.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("/workspace/output/sorted_products")

print("Sorted products saved")
```

```python
top3_df = sorted_df.limit(3)

print("===== Top 3 Highest Selling Products =====")

top3_df.show()
top3_df.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("/workspace/output/top_3_products")

print("Top 3 products saved")
```

```python
high_sales_df = df.filter(
     col(" sales ") > 80000
)

print("===== Products With Sales Greater Than 80000 =====")

high_sales_df.show()
high_sales_df.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("/workspace/output/high_sales_products")

print("High sales products saved")
```
