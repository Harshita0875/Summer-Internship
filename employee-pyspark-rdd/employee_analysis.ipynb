```python
from pyspark.sql import SparkSession

# Create Spark Session
spark = SparkSession.builder \
    .appName("EmployeeRDDAnalysis") \
    .master("local[*]") \
    .getOrCreate()

sc = spark.sparkContext
```

```python
rdd = sc.textFile("data/employee.csv")

header = rdd.first()

employees = rdd.filter(lambda x: x != header) \
               .map(lambda x: x.split(","))
```

```python
sorted_output = [
    f"{emp[0]},{emp[1]},{emp[2]},{emp[3]}"
    for emp in sorted_employees.collect()
]


with open("output/sorted_employees.txt", "w") as f:
    f.write("id,name,department,salary\n")
    f.write("\n".join(sorted_output))


print("\nSorted employee data saved to sorted_employees.txt")
```

```python
dept_salary = employees.map(
    lambda x: (x[2], int(x[3]))
)

dept_totals = dept_salary.reduceByKey(
    lambda a, b: a + b
)


print("\n===== Department Salary Totals =====")

for dept, total in dept_totals.collect():
    print(f"{dept}: {total}")


# Save department salary totals to file

dept_output = [
    f"{dept},{total}"
    for dept, total in dept_totals.collect()
]


with open("output/department_salary_totals.txt", "w") as f:
    f.write("department,total_salary\n")
    f.write("\n".join(dept_output))


print("\nDepartment salary totals saved to department_salary_totals.txt")
```

```python
top3 = sorted_employees.take(3)

output_lines = [
    f"{emp[0]},{emp[1]},{emp[2]},{emp[3]}"
    for emp in top3
]

with open("output/top_3_employees.txt", "w") as f:
    f.write("id,name,department,salary\n")
    f.write("\n".join(output_lines))

print("\nTop 3 employees saved to top_3_employees.txt")
```

```python
spark.stop()
```
