```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PartitionManagement") \
    .master("local[*]") \
    .getOrCreate()

print("Spark Version:", spark.version)
```

```python
df = spark.range(5000000)

print("Total Records:", df.count())
```

```python
initial_partitions = df.rdd.getNumPartitions()

print("Initial Number of Partitions:", initial_partitions)
```

```python
df_repartitioned = df.repartition(15)

print(
    "Partitions after repartition(15):",
    df_repartitioned.rdd.getNumPartitions()
)
```

```python
df_coalesced = df_repartitioned.coalesce(3)

print(
    "Partitions after coalesce(3):",
    df_coalesced.rdd.getNumPartitions()
)
```

```python
df_coalesced.show(10)
```

```python
import os

os.makedirs("/workspace/output", exist_ok=True)

with open("/workspace/output/partition_results.txt", "w") as f:
    f.write(f"Initial Partitions: {initial_partitions}\n")
    f.write(f"After Repartition(12): {df_repartitioned.rdd.getNumPartitions()}\n")
    f.write(f"After Coalesce(3): {df_coalesced.rdd.getNumPartitions()}\n")

print("Results saved successfully.")
```
